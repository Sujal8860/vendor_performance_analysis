```python
import time

start = time.time()

purchase_summary = pd.read_sql_query("""
SELECT 
    p.VendorNumber,
    p.VendorName,
    p.Brand,
    p.Description,
    p.PurchasePrice,
    pp.Price as ActualPrice,
    pp.Volume,
    SUM(p.Quantity) AS TotalPurchaseQuantity,
    SUM(p.Dollars) AS TotalPurchaseDollars
FROM purchases p
JOIN purchase_prices pp
    ON p.Brand = pp.Brand
WHERE p.PurchasePrice > 0
GROUP BY p.VendorNumber, p.VendorName, p.Brand, p.Description, p.PurchasePrice, pp.Price, pp.Volume
ORDER BY TotalPurchaseDollars
""", con)

end = time.time()

print(f"â± Query executed in {(end - start):.2f} seconds")
purchase_summary
```

```python
start = time.time()
sales_summary = pd.read_sql_query("""
SELECT 
    VendorNo,
    Brand,
    SUM(SalesQuantity) AS TotalSalesQuantity,
    SUM(SalesDollars) AS TotalSalesRevenue,  -- Corrected this line
    AVG(SalesPrice) AS AvgSalesPrice,         -- Assuming we want the average sales price per unit
    SUM(ExciseTax) AS TotalExciseTax
FROM sales
GROUP BY VendorNo, Brand
""", con)

end = time.time()
print(f"excuted time : {(end-start):.2f} seconds")
sales_summary
```

```python
# merging all the summary table to create an aggregated summary table for further analysis
start = time.time()
vendor_sales_summary = pd.read_sql_query("""WITH FreightSummary AS (
    SELECT 
        VendorNumber, 
        SUM(Freight) AS FreightCost 
    FROM vendor_invoice 
    GROUP BY VendorNumber
), 

PurchaseSummary AS (
    SELECT 
        p.VendorNumber,
        p.VendorName,
        p.Brand,
        p.Description,
        p.PurchasePrice,
        pp.Price AS ActualPrice,
        pp.Volume,
        SUM(p.Quantity) AS TotalPurchaseQuantity,
        SUM(p.Dollars) AS TotalPurchaseDollars
    FROM purchases p
    JOIN purchase_prices pp
        ON p.Brand = pp.Brand
    WHERE p.PurchasePrice > 0
    GROUP BY p.VendorNumber, p.VendorName, p.Brand, p.Description, p.PurchasePrice, pp.Price, pp.Volume
), 

SalesSummary AS (
    SELECT 
        VendorNo,
        Brand,
        SUM(SalesQuantity) AS TotalSalesQuantity,
        SUM(SalesDollars) AS TotalSalesDollars,
        SUM(SalesPrice) AS TotalSalesPrice,
        SUM(ExciseTax) AS TotalExciseTax
    FROM sales
    GROUP BY VendorNo, Brand
) 

SELECT 
    ps.VendorNumber,
    ps.VendorName,
    ps.Brand,
    ps.Description,
    ps.PurchasePrice,
    ps.ActualPrice,
    ps.Volume,
    ps.TotalPurchaseQuantity,
    ps.TotalPurchaseDollars,
    ss.TotalSalesQuantity,
    ss.TotalSalesDollars,
    ss.TotalSalesPrice,
    ss.TotalExciseTax,
    fs.FreightCost
FROM PurchaseSummary ps
LEFT JOIN SalesSummary ss 
    ON ps.VendorNumber = ss.VendorNo 
    AND ps.Brand = ss.Brand
LEFT JOIN FreightSummary fs 
    ON ps.VendorNumber = fs.VendorNumber
ORDER BY ps.TotalPurchaseDollars DESC""",con)
end = time.time()
print(f"excuted time : {(end-start):.2f} seconds")
vendor_sales_summary.head()
```

âš™ï¸ Efficiency Improvement:

-It performs complex joins and aggregations across large tables like sales and purchases.

- Saving the summarized output helps eliminate repeated heavy processing.

- Useful for understanding vendor and brand-wise trends in sales, purchase volume, and pricing.

- Beneficial for future use in reporting or dashboards by reducing load times.

- Instead of executing the same heavy query repeatedly, dashboards can instantly pull data from vendor_sales_summary.

# Data Cleaning

```python
vendor_sales_summary.info()
```

there are some columns which contain some null values and it can be possible because the vendor purchases some item but didn't sell it that's why the value remains  null

```python
# Fill nulls with 0 (only in numeric columns)
vendor_sales_summary.fillna(0, inplace=True)
#confirming
vendor_sales_summary.isnull().sum()

```

```python
vendor_sales_summary.duplicated().sum()  # Count duplicates
```

```python
vendor_sales_summary['VendorName'].unique()
```

```python
#There are white spaces in the categorical columns.
vendor_sales_summary.columns = vendor_sales_summary.columns.str.strip().str.lower().str.replace(' ', '_')
vendor_sales_summary['vendorname'] = vendor_sales_summary['vendorname'].str.replace(r'\s+', ' ', regex=True).str.strip()
vendor_sales_summary.head()
```

```python
vendor_sales_summary['totalsalesdollars'].describe()
```

```python
vendor_sales_summary['volume'] = vendor_sales_summary['volume'].astype('float64')
```

```python
# âž• Creating new analytical columns with error handling (e.g. avoid division by zero)
vendor_sales_summary['gross_profit'] = vendor_sales_summary['totalsalesdollars'] - vendor_sales_summary['totalpurchasedollars']

vendor_sales_summary['profit_margin'] = vendor_sales_summary.apply(
    lambda x: (x['gross_profit'] / x['totalsalesdollars'] * 100) if x['totalsalesdollars'] != 0 else 0,
    axis=1
)

vendor_sales_summary['stock_turnover'] = vendor_sales_summary.apply(
    lambda x: (x['totalsalesquantity'] / x['totalpurchasequantity']) if x['totalpurchasequantity'] != 0 else 0,
    axis=1
)

vendor_sales_summary['sales_to_purchase_ratio'] = vendor_sales_summary.apply(
    lambda x: (x['totalsalesdollars'] / x['totalpurchasedollars']) if x['totalpurchasedollars'] != 0 else 0,
    axis=1
)
```

### saving clean data to db

```python
cursor = con.cursor()

cursor.execute("""
CREATE TABLE vendor_sales_summary (
    vendor_number INT,
    vendor_name VARCHAR(100),
    brand INT,
    description VARCHAR(100),
    purchase_price DECIMAL(10,2),
    actual_price DECIMAL(10,2),
    volume INT,
    totalpurchasequantity INT,
    totalpurchasedollars DECIMAL(15,2),
    totalsalesquantity INT,
    totalsalesdollars DECIMAL(15,2),
    totalsalesprice DECIMAL(15,2),
    totalexcisetax DECIMAL(15,2),
    freightcost DECIMAL(15,2),
    gross_profit DECIMAL(15,2),
    profit_margin DECIMAL(15,2),
    stock_turnover DECIMAL(15,2),
    sales_to_purchase_ratio DECIMAL(15,2),
    PRIMARY KEY (vendor_number, brand)
);
""")
```
